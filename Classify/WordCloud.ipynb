## 生成词云图

```python
# 库导入
import os
import jieba
from PIL import Image
from matplotlib import pyplot as plt
from wordcloud import WordCloud
import numpy as np
import pandas as pd
```

```python
# 读取词频文件
df=pd.read_csv("./word_freq.csv",encoding='gbk')
df.head(10)
```

```python
text = (open('./word_freq.csv','r',encoding='gbk')).read()

wordcloudfigure = WordCloud(
    background_color='white',
    width=1024,
    height=1024,
    font_path=os.getenv('SystemRoot') + '\\Fonts\\simsun.ttc', # 电脑自带的字体
    max_words=500, # 设置最大现显示的字数
    stopwords=['word','freq'], # 设置停止词
)

wordcloudfigure.generate_from_text(text) # 绘制图片
plt.imshow(wordcloudfigure)
plt.axis('off')
plt.show() # 显示图片
wordcloudfigure.to_file('./wordcloud.png')
```
